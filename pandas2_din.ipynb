```python
hr.loc[0: 99, 'Employee ID'] = ['E' +  str(i).zfill(4)
                                for i in range(1, len(hr) + 1)]
hr
```

```python
hr['Department'] = hr['Department'].fillna('Data Analyst')
hr
```

```python
hr['Hire Date'] = hr['Hire Date'].interpolate(method='linear')
hr
```

```python
hr.loc[[14, 23, 28, 40, 66, 79, 81], 'Current Position'] = ['Mid', 'Manager', 'Junior', 'Lead', 'Junior', 'Mid', 'Senior']
hr
```

```python
hr.loc[83,'Salary (USD)'] = 2000

hr.loc[hr['Salary (USD)'] < 0, 'Salary (USD)'] = hr['Salary (USD)'].abs()
hr.loc[hr['Salary (USD)'] < 11, 'Salary (USD)'] = hr.loc[hr['Salary (USD)'] < 11, 'Salary (USD)'] * 100

hr
```
