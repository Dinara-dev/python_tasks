```python
# 3. GroupBy bilan ishlash
# Category bo‘yicha o‘rtacha narxni hisoblang.

polo = pd.read_csv('foods.csv')
category = polo.groupby('Category')['Price'].mean()
category
```

```python
# Place bo‘yicha jami buyurtma sonini toping.
polo = polo.groupby('Place')['Quantity'].count()
```

```python
# Har bir Customer bo‘yicha sarflagan umumiy summani (Price * Quantity) hisoblang
polo['umumiy_summa'] = polo['Price'] * polo['Quantity']
polo.groupby('Customer')['umumiy_summa'].sum()
```

```python
# 4. Pivot Table
# Place va Category bo‘yicha pivot jadval tuzing, unda qiymat sifatida Quantity yig‘indisi chiqsin.

pivot = pd.pivot_table(polo, values = 'Quantity', index = 'Place', columns = 'Category',
                       aggfunc = 'sum')
pivot
```

```python
# Customer va Food bo‘yicha pivot_table tuzing va o‘rtacha Ratingni ko‘rsating.
pivot = pd.pivot_table(polo, values = 'Rating', index = 'Customer', columns = 'Food',
                       aggfunc = 'mean')
pivot
```

```python
# 5. Apply va qo‘shimcha hisoblash
# apply yordamida yangi ustun Total yarating (Price * Quantity).

df = pd.read_csv('foods.csv')

def count_total(row):
    price = row['Price']
    quantity = row['Quantity']
    total = price * quantity
    return total
df['Total'] = df.apply(count_total, axis=1)

print(df)
```

```python
# Rating ustuniga funksiya qo‘llab, agar baho 5 bo‘lsa "Excellent", 3-4 bo‘lsa "Good", 1-2 bo‘lsa "Bad" deb yozing.

def assign_grade(Rating):
    if Rating == 5:
        return 'Excellent'
    elif Rating == 4 or Rating == 3:
        return 'Good'
    elif Rating == 1 or Rating == 2:
        return 'Bad'
    else:
        return 'Do not found'

polo['Grade'] = polo['Rating'].apply(assign_grade)
polo[['Food', 'Rating', 'Grade']]
```

```python
# Oxirgi 7 kunda berilgan buyurtmalar sonini hisoblang.

df = pd.read_csv('foods.csv')

df['Date'] = pd.date_range(start='2026-02-15', periods=len(df), freq='D')
df["Date"] = pd.to_datetime(df["Date"])
today = pd.to_datetime('today')
seven_days_ago = today - pd.Timedelta(days=7)
last_week_orders = df[df['Date'] >= seven_days_ago]
num_orders = len(last_week_orders)

print(f'Oxirgi 7 kunda berilgan buyurtmalar soni: {num_orders}')
print("Buyurtmalar ro'yxati:")
print(last_week_orders[['Food', 'Date']])



```

```python
# Har bir oy bo‘yicha jami tushumni (Price * Quantity) toping.

df = pd.read_csv('foods.csv')

df['Date'] = pd.date_range(start='2026-02-15', periods=len(df), freq='D')
df['Date'] = pd.to_datetime(df['Date'])
df['Total'] = df['Price'] * df['Quantity']
df['Month'] = df['Date'].dt.to_period('M')
monthly_income = df.groupby('Month')['Total'].sum()

print(monthly_income)
```

```python
# 7. GroupBy + Apply
# Har bir Customer uchun o‘rtacha chek summasini hisoblang.

df = pd.read_csv('foods.csv')
df['Total'] = df['Price'] * df['Quantity']

def customer_summary(group):
    avg_total = group['Total'].mean()
    group = group.copy()
    group['Avg_Check'] = avg_total
    return group[['Food', 'Price', 'Quantity', 'Total', 'Avg_Check']]
result = df.groupby('Customer').apply(customer_summary)

result = df.groupby('Customer', group_keys=False).apply(customer_summary)

print(result)
```

```python
# Quantity >= 5 bo‘lgan buyurtmalardan jami tushumni hisoblang.

df = pd.read_csv('foods.csv')

filtered = df.loc[df['Quantity'] >= 5].copy()
filtered.loc[:, 'Revenue'] = filtered['Price'] * filtered['Quantity']
total_revenue = sum(filtered['Revenue'])

print('Quantity >= 5 bo‘lgan buyurtmalardan jami tushum:', total_revenue)
```

```python
# Customer bo‘yicha jami Totalni hisoblang va eng ko‘p pul sarflagan mijozni aniqlang.

df = pd.read_csv('foods.csv')


df['Total'] = df['Price'] * df['Quantity']
customer_total = df.groupby('Customer')['Total'].sum()
top_customer = customer_total.idxmax()
top_amount = customer_total.max()

print("Eng ko‘p pul sarflagan mijoz:", top_customer, '\n', 'Umumiy sarfi:', top_amount)
```

```python
# Sana bo‘yicha rolling(7) yordamida haftalik tushumni ko‘rsating.

df = pd.read_csv('foods.csv')

df['Date'] = pd.date_range(start='2026-03-01', periods=len(df), freq='D')
df['Total'] = df['Price'] * df['Quantity']
df = df.sort_values('Date')
df['Weekly_Revenue'] = df['Total'].rolling(7).sum()

print(df[['Date','Total','Weekly_Revenue']])
```
