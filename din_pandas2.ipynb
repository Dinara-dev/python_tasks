```python
import pandas as pd

hr = pd.read_excel('HR_Dataset_for_PowerBI_with_Errors.xlsx')
hr
```

```python

hr.loc[0: 99, 'Employee ID'] = ['E' + str(i).zfill(4)
                                for i in range(1, 101)]
hr
```

```python
hr.isnull().sum()
```

```python
hr['Department'] = hr['Department'].fillna('Data Analyst')
hr
```

```python
# hr['Hire Date'] = hr['Hire Date'].fillna(hr['Hire Date'].mean()).dt.date

hr['Hire Date'] = pd.to_datetime(hr['Hire Date'])
hr['Hire Date'] = hr['Hire Date'].interpolate()
hr
```

```python
hr.loc[14,'Current Position'] = 'Mid'
hr
```

```python
hr.loc[[16,23,28], 'Current Position'] = ['Mid', 'Senior', 'Junior']
hr
```

```python
hr.loc[[40,66,79,81], 'Current Position'] = ['Manager', 'Lead', 'Junior', 'Mid']
hr
```

```python
hr.loc[83,'Salary (USD)'] = 2000

hr.loc[hr['Salary (USD)'] < 0, 'Salary (USD)'] = hr['Salary (USD)'].abs()
hr['Salary (USD)'] = hr['Salary (USD)'].round()
hr
```
